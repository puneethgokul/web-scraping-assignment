## PART1

```python
from bs4 import BeautifulSoup
import requests
```

```python
page1 = requests.get('https://www.icc-cricket.com/rankings/womens/team-rankings/odi')
```

```python
page1.content
```

```python
soup1 = BeautifulSoup(page1.content, 'html.parser')

print(soup1.prettify())
```

Grab Teams

```python
topteam = soup1.find_all('td', class_="rankings-block__banner--team-name")
topteam
```

```python
topteam[0].find('span', class_="u-hide-phablet").text
```

```python
top10ODITEAMS = []

top10ODITEAMS.append(topteam[0].find('span', class_="u-hide-phablet").text)
```

```python
twototenteams = soup1.find_all('td', class_="table-body__cell rankings-table__team")
twototenteams[0:4]
```

```python
twototenteams[0].find('span', class_="u-hide-phablet").text
```

```python
for i in twototenteams[0:9]:
    top10ODITEAMS.append(i.find('span', class_="u-hide-phablet").text)
    
top10ODITEAMS
```

Grab Matches

```python
topteammatches = soup1.find_all('td', class_="rankings-block__banner--matches")
topteammatches
```

```python
topteammatches[0].text
```

```python
top10ODITEAMSMatches = []

top10ODITEAMSMatches.append(topteammatches[0].text)
```

```python
twototenteamsMatches = soup1.find_all('td', class_="table-body__cell u-center-text")
twototenteamsMatches[0:4]
```

```python
twototenteamsMatches[0].text
```

```python
for i in twototenteamsMatches[0:18]:
    if twototenteamsMatches.index(i)%2==0:
         top10ODITEAMSMatches.append(i.text)
   
    
top10ODITEAMSMatches
```

Grab Points

```python
topteamPoints = soup1.find_all('td', class_="rankings-block__banner--points")
topteamPoints
```

```python
topteamPoints[0].text
```

```python
top10ODITEAMSPoints = []

top10ODITEAMSPoints.append(topteamPoints[0].text)
```

```python
twototenteamsPoints = soup1.find_all('td', class_="table-body__cell u-center-text")
twototenteamsPoints[0:4]
```

```python
twototenteamsPoints[1].text
```

```python
for i in twototenteamsPoints[0:18]:
    if twototenteamsPoints.index(i)%2==1:
         top10ODITEAMSPoints.append(i.text)
   
    
top10ODITEAMSPoints
```

Grap Ratings

```python
topteamRatings = soup1.find_all('td', class_="rankings-block__banner--rating u-text-right")
topteamRatings
```

```python
topteamRatings[0].text.replace('\n','').strip()
```

```python
top10ODITEAMSRatings = []

top10ODITEAMSRatings.append(topteamRatings[0].text.replace('\n','').strip())
```

```python
twototenteamsRatings = soup1.find_all('td', class_="table-body__cell u-text-right rating")
twototenteamsRatings[0:4]
```

```python
twototenteamsRatings[1].text
```

```python
for i in twototenteamsRatings[0:9]:
         top10ODITEAMSRatings.append(i.text)
   
    
top10ODITEAMSRatings
```

create final data frame

```python
import pandas as pd

top10ODImens = pd.DataFrame()
```

```python
top10ODImens['top10ODITEAMS'] = top10ODITEAMS
top10ODImens['top10ODITEAMSMatches'] = top10ODITEAMSMatches
top10ODImens['top10ODITEAMSPoints'] = top10ODITEAMSPoints
top10ODImens['top10ODITEAMSRatings'] = top10ODITEAMSRatings
```

```python
top10ODImens
```
