```python
from bs4 import BeautifulSoup
import requests
```

```python
page1 = requests.get('https://en.wikipedia.org/wiki/Main_Page')
```

```python
page1.content
```

```python
soup1 = BeautifulSoup(page1.content, 'html.parser')
```

```python
print(soup1.prettify())
```

```python
headers = soup1.find_all('span', class_="mw-headline")
```

```python
headers
```

```python
for i in headers:
    print(i.get_text())
```
